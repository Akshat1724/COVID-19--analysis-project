```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []

for c in countries:
    # Safely convert to numeric before diff
    data = pd.to_numeric(corona_dataset_aggregated.loc[c], errors='coerce')
    max_diff = data.diff().max()
    max_infection_rates.append(max_diff)

corona_dataset_aggregated["Max_Infection_Rates"] = max_infection_rates

```

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["Max_Infection_Rates"])
corona_data
```
